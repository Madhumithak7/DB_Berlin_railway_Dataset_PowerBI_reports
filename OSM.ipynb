```python
import osmnx as ox

# Define the area of interest (e.g., Berlin, Germany)
place = "Berlin, Germany"

# Fetch railway tracks
railway_tracks = ox.features.features_from_place(place, tags={"railway": "halt"})

# Fetch railway stations
railway_stations = ox.features.features_from_place(place, tags={"railway": "switch"})

# Fetch railway signals
railway_signals = ox.features.features_from_place(place, tags={"railway": "crossing"})

# Save the data to GeoJSON files
railway_tracks.to_file("berlin_railway_halt.geojson", driver="GeoJSON")
railway_stations.to_file("berlin_railway_switch.geojson", driver="GeoJSON")
railway_signals.to_file("berlin_railway_crossing.geojson", driver="GeoJSON")

print("Data extraction complete!")
```
